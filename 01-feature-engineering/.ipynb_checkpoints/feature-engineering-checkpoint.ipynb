## 数据读取

```python
# -*- coding: utf-8 -*-

# 加载需要的库
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 载入数据
train = pd.read_csv(u'train.csv')
test = pd.read_csv(u'test.csv')
train.shape,test.shape
```

##  查看数据基本情况

```python
train.dtypes
```

```python
# 拿出前五条
train.head(5)
```

```python
# train和test合成
train['source']= 'train'
test['source'] = 'test'
data=pd.concat([train, test],ignore_index=True)
data.shape
```

```python
data.tail(5)
```

## 查看数据高级用法

```python
# 看看异常点，比如说缺省值
# data.apply(lambda x: sum(x.isnull()))
count_null = lambda x: sum(x.isnull())
data.apply(count_null)
```

```python
# 看看这些字段，分别有多少种取值(甚至你可以看看分布)
var = ['Embarked','Sex','SibSp','Survived','Ticket']
for v in var:
    print('\n%s这一列数据的不同取值和出现的次数\n'%v)
    print(data[v].value_counts())
```
